```python
import requests
import time
import matplotlib.pyplot as plt

cache = dict()

def extract_article_content(url):
    response = requests.get(url)
    content = response.content
    return content

def action(url):
    if url not in cache:
        content = extract_article_content(url)
        cache[url] = content

    return cache[url]

N=10
timings=[]
for i in range(N):
    start = time.time()
    action("https://translate.google.com/")
    end = time.time()
    timings.append(end-start)
    start=end
    
plt.plot(timings,'r',linewidth=4) 
plt.xlabel('computation#01')
plt.ylabel('seconds')
plt.grid(True)
```
